```python
import pandas as pd
import numpy
import csv

stock_data = pd.read_csv('./거시경제지표.csv',encoding='cp949')

date_all = stock_data['Date']

date_all
```

```python
# 관련 라이브러리를 호출합니다.
import requests

# 관련 라이브러리를 호출합니다.
from bs4 import BeautifulSoup as bts

import re

title_text=[]
date_text=[]
context_text=[]
result={}

for i in date_all:
    
    # 요청 URL을 설정합니다.
    url = 'https://finance.naver.com/news/mainnews.nhn'

    # 오늘 날짜를 yyyy-mm-dd 형태의 문자열로 생성합니다.
    today = i


    # Query String을 딕셔너리로 생성합니다.
    query = {'date': today}

    # HTTP 요청을 실행합니다.
    res = requests.get(url = url, params = query)


    # str 자료형을 bs4.BeautifulSoup 자료형으로 변환합니다.
    soup = bts(markup = res.text, features = 'html.parser')


    # 뉴스 제목을 포함하는 HTML 요소만 선택하고 items에 할당합니다.
    title = soup.select(selector = 'ul.newsList > li')


    title1 = soup.select('dt > a')
    title2 = soup.select('dd > a')
    context = soup.select('dd.articleSummary')
    date = soup.select('dd.articleSummary > span.wdate')
    
    k=0
    for j in range(len(title1)):
        if(title1[j].text == ""):
            title_text.append(title2[k].text)
            k += 1
        else:
            title_text.append(title1[j].text)

        date_text.append(date[j].text)
        context_text.append(re.sub('\t|\r|\n',"", context[j].text.split("..")[0]))
        
#     print(title_text , date_text, context_text)
 
   
        
    num = soup.select('div#wrap > div#newarea > div#contentarea > div#contentarea_left > table > tr > td > a ')

    if(len(num)!=1):
        for j in range(len(num)-2):
            url = f'https://finance.naver.com/news/mainnews.nhn?date={i}&page={j+2}'
            
            # HTTP 요청을 실행합니다.
            res = requests.get(url = url)


            # str 자료형을 bs4.BeautifulSoup 자료형으로 변환합니다.
            soup = bts(markup = res.text, features = 'html.parser')


            # 뉴스 제목을 포함하는 HTML 요소만 선택하고 items에 할당합니다.
            title = soup.select(selector = 'ul.newsList > li')


            title1 = soup.select('dt > a')
            title2 = soup.select('dd > a')
            context = soup.select('dd.articleSummary')
            date = soup.select('dd.articleSummary > span.wdate')

            k=0
            for j in range(len(title1)):
                if(title1[j].text == ""):
                    title_text.append(title2[k].text)
                    k += 1
                else:
                    title_text.append(title1[j].text)

                date_text.append(date[j].text)
                context_text.append(re.sub('\t|\r|\n',"", context[j].text.split("..")[0]))
        
    print(title_text , date_text, context_text)
                
    #모든 리스트 딕셔너리형태로 저장
    # id= date , document = title , label 은 일단 3으로 고정(의미없는 값)
    result= {"id" : date_text , "title":title_text , "context":context_text }  


    df = pd.DataFrame(result)  #df로 변환

#     인덱스 없애기
    df.to_csv(f'main_news_title_time.txt', sep = '\t', index = False)

```
